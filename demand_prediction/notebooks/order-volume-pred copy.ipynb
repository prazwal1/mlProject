When analysing the disccount construct column it was found that aside from the percentage of discount, the flat offers were already extracted in the restaurant discount(promo) column. So we are extracting percentage and dropping the construct column entirely.

```python
import re

# Extract percentage from Discount construct
def extract_percent(x):
    if pd.isna(x):
        return 0
    match = re.search(r"(\d+)%", x)
    return float(match.group(1)) if match else 0

df["discount_percent"] = df["Discount construct"].apply(extract_percent)

# Drop the original textual column
df = df.drop(columns=["Discount construct"])
```

```python
# Remove 'km' and handle <1km
def convert_distance(x):
    if pd.isna(x):
        return None
    x = x.replace("km", "").strip()
    if x.startswith("<"):
        return 0.9  # or any small number for <1km
    try:
        return float(x)
    except:
        return None

df["Distance"] = df["Distance"].apply(convert_distance)
```

```python
import re

def extract_order_count(items):
    if pd.isna(items):
        return 0
    # Match the first number before 'x'
    match = re.match(r'(\d+)\s*x', items.strip())
    if match:
        return int(match.group(1))
    return 1  # default to 1 if no number is found
```

```python
df['order_count'] = df['Items in order'].apply(extract_order_count)
df['order_count'].value_counts()
```

## Feature Engineering Our Target Variable: Order Count

```python

# Aggregate to get orders per day per hour
orders_hourly = df.groupby(['order_date', 'order_hour'])['order_count'].sum().reset_index(name='orders_per_hour')

# Aggregate to get orders per day (total)
orders_daily = df.groupby('order_date')['order_count'].sum().reset_index(name='orders_per_day')

# Sort by date and hour to ensure correct order for lag/rolling features
orders_hourly = orders_hourly.sort_values(by=['order_date', 'order_hour'])

# Create Lag Features (e.g., orders from 1 hour ago and 24 hours ago)
orders_hourly['orders_lag_1hr'] = orders_hourly['orders_per_hour'].shift(1)
orders_hourly['orders_lag_24hr'] = orders_hourly['orders_per_hour'].shift(24) # Same hour, previous day

# First, shift the data by 1 hour, so we are only looking at PAST data.
# Then, apply the rolling window on this shifted data.
shifted_orders = orders_hourly['orders_per_hour'].shift(1)

# Now the window at time 't' will be looking at data from t-1, t-2, and t-3
orders_hourly['rolling_mean_3hr'] = shifted_orders.rolling(window=3, min_periods=1).mean()
orders_hourly['rolling_max_3hr'] = shifted_orders.rolling(window=3, min_periods=1).max()

# Fill any initial NaN values that result from shifting
orders_hourly.fillna(0, inplace=True)
```

```python
# Preprocessing & evaluation
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
from sklearn.preprocessing import StandardScaler

# --- Time-Based Split ---
# Ensure the data is sorted chronologically before splitting
ml_df = ml_df.sort_values(by=['order_date', 'order_hour'])

# Define a cutoff date for the split
# For example, use the last 10% of the data for testing
split_index = int(len(ml_df) * 0.8)
train_df = ml_df.iloc[:split_index]
test_df = ml_df.iloc[split_index:]

# Define features and target
target = 'orders_per_hour'
features_to_drop = ['order_date', 'orders_per_hour', 'Restaurant name', 'Restaurant ID', 'order_hour']
X_train = train_df.drop(columns=features_to_drop)
y_train = train_df[target]

X_test = test_df.drop(columns=features_to_drop)
y_test = test_df[target]

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

```python
# 1. Initialize the Scaler
scaler = StandardScaler()

# 2. Create copies of the dataframes to hold the scaled values
#    This keeps your original X_train and X_test for the tree-based models
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

# 3. Fit the scaler only on the training data and transform it
X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])

# 4. Use the SAME scaler (already fitted) to transform the test data
X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])

print(X_train_scaled.columns)
```

```python
# --- Comprehensive Model Tuning and Evaluation ---

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit

# 1. Define Models and Hyperparameter Grids to Search
# These are example grids. You can expand them, but be aware it will take longer.
models_and_params = {
    'RandomForest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'XGBoost': {
        'model': XGBRegressor(random_state=42, objective='reg:squarederror'),
        'params': {
            'n_estimators': [100, 300],
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5]
        }
    },
    'NeuralNet': {
        'model': MLPRegressor(random_state=42, max_iter=500, early_stopping=True),
        'params': {
            'hidden_layer_sizes': [(50,), (100, 50)],
            'activation': ['relu', 'tanh'],
            'alpha': [0.0001, 0.001]
        }
    }
}

# 2. Prepare for a loop to tune and evaluate each model
results = {}
# Use TimeSeriesSplit for cross-validation in our grid search
tscv = TimeSeriesSplit(n_splits=4)

for name, mp in models_and_params.items():
    print(f"--- Tuning model: {name} ---")

    # Use scaled data ONLY for the Neural Network
    is_neural_net = (name == 'NeuralNet')
    X_train_data = X_train_scaled if is_neural_net else X_train
    X_test_data = X_test_scaled if is_neural_net else X_test

    # Set up and run GridSearchCV
    grid_search = GridSearchCV(estimator=mp['model'],
                               param_grid=mp['params'],
                               cv=tscv,
                               scoring='neg_mean_squared_error',
                               verbose=1,
                               n_jobs=-1)
    grid_search.fit(X_train_data, y_train)

    # Get the best model from the search
    best_model = grid_search.best_estimator_
    print(f"Best parameters for {name}: {grid_search.best_params_}")

    # Evaluate the best model on the test data
    y_pred = best_model.predict(X_test_data)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred)**0.5
    mape = mean_absolute_percentage_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    # Store the results
    results[name] = {'MAE': mae, 'RMSE': rmse, 'MAPE': mape, 'R2 Score': r2}
    print(f"Evaluation for tuned {name} complete.\n")

# 3. Display the final comparative results
results_df = pd.DataFrame(results).T
print("--- Final Model Comparison (After Tuning) ---")
print(results_df)
```
